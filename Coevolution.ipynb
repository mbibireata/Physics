```python
N = 20000 #number of iterations
T = 100.0 #time 
dt = T/N #step size 

#make a competition matrix
m = 10 #pairs of prey/predators 
e = zeros((m,m))

for i in range (m):
    for j in range(i+1,m): 
        e[i][j] = uniform(.05,.15)#.1
        e[j][i] = e[i][j]

prey = zeros((N,m))
pred = zeros((N,m))

#initialize with some populations
prey[0] = ones(10)*.5
pred[0] = ones(10)*.5

b_rate = zeros(m) #birth rate
p_rate = zeros(m) #predation rate
d_rate = zeros(m) #decay rate
B_size = zeros(m) #burst size

for i in range (m):
    b_rate = [.75, .8, .85, .9, .95, 1, 1.05, 1.1, 1.15, 1.2]
    p_rate[i] = 2
    d_rate[i] = .5
    B_size[i] = 1

for i in range (N - 1):
    for j in range(m): 
        prey[i + 1][j] = prey[i][j] + dt*(b_rate[j]*prey[i][j] - p_rate[j]*pred[i][j]*prey[i][j] - prey[i][j]*dot(e[j],prey[i])) 
        #if(prey[i + 1][j]<=0): 
            #prey[i + 1][j] = 0
        
            
        pred[i + 1][j] = pred[i][j] + dt*(B_size[j]*p_rate[j]*prey[i][j]*pred[i][j] - d_rate[j]*pred[i][j])
        #if (pred[i + 1][j] <= 0): 
            #pred[i + 1][j] = 0
        
figure(1)
title("Prey")
xlabel("time")
ylabel("population density")
for i in range(m):
    plot(prey.transpose()[i])


figure(2)
title("Predators")
xlabel("time")
ylabel("population density")
for i in range(m):
    plot(pred.transpose()[i])
    
    



```
